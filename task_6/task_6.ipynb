```python
import random

OBSTACLES = 40

def get_new_node(new_map: list):
    x = random.randint(0,9)
    y = random.randint(0,19)
    
    while new_map[x][y] == 'X':
        x = random.randint(0,9)
        y = random.randint(0,19)

    return (x,y)


def generate_new_map(size, n_blocks):
    new_map = [[0] * size[1] for _ in range(size[0])]
    for i in range(n_blocks):
        x, y = get_new_node(new_map)
        new_map[x][y] = 'X'
    
    return new_map
```

```python
def get_neighbor_vertexes(i, j):
  neighbors = []
  if i - 1 >= 0 and new_map[i-1][j] != 'X':
    neighbors.append((i-1, j))
  if j - 1 >= 0 and new_map[i][j-1] != 'X':
    neighbors.append((i, j-1))
  if j + 1 < map_size[1] and new_map[i][j + 1] != 'X':
    neighbors.append((i, j+1))
  if i + 1 < map_size[0] and new_map[i+1][j] != 'X':
    neighbors.append((i+1, j))
  return neighbors
```

```python
def get_node_index(i, j):
  return i * 19 + j + 1 * (i > 0)
```

```python
def transform_map_to_adj_matrix(new_map, map_size):
  graph_vertexes = []
  for i in range(map_size[0]):
    for j in range(map_size[1]):
      if new_map[i][j] != 'X':
        neighbors = get_neighbor_vertexes(i, j)
        for neighbor in neighbors:
          graph_vertexes.append((get_node_index(i, j), get_node_index(*neighbor), 1))
  return graph_vertexes
```
